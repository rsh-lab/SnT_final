```python
%pip install yahoofinancials
%pip install yfinance
%pip install seaborn
import yfinance as yf
import seaborn as sns
from yahoofinancials import YahooFinancials
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
yahoo_financials = YahooFinancials('LT.NS')
data = yahoo_financials.get_historical_price_data(start_date='2022-08-1', 
                                                  end_date='2023-03-31', 
                                                  time_interval='daily')
lt_df = pd.DataFrame(data['LT.NS']['prices'])
lt_df = lt_df.drop('date', axis=1).set_index('formatted_date')
lt_df.head()
```

```python
sma=[lt_df['close'][i-50:i].mean() for i in range (75, 165)]
```

```python
sma=[np.nan]*75+sma
```

```python
from matplotlib import pyplot as plt
```

```python
ewm1=lt_df['close'].ewm(span = 50, adjust = False).mean() #EMA using direct function
```

```python
ewm1
lt_df['ewm']=ewm1
```

```python
lt_df.index
```

```python
plt.plot(lt_df.index,lt_df['sma'], label='SMA')
plt.plot(lt_df.index,lt_df['ewm'], label='EMA')
plt.legend(loc='upper left')
```
